```python
from pandasql import sqldf
q = lambda query: sqldf(query, globals()) # Shortcut to run SQL
```

```python
a1 = """
SELECT
    t.user_id AS customer_id,
    u.country AS customer_country,
    t.currency,
    t.amount AS transaction_amount
FROM transactions t
LEFT JOIN users u
    ON t.user_id = u.id
WHERE t.source = 'GAIA'
"""
```

```python
a1_result = q(a1)
a1_result.head()
```
